```python
mean = [0,0]
cov = [[3,3],[3,4]] # para darle el grado extra de libertad y que los datos salgan mas dispersos creo hay que darle una varianza a la combinacion de coordenadas 
size = 500
X = np.random.multivariate_normal(mean=mean, cov=cov, size=size)
```

```python
X
```

```python
plt.scatter(X[:,0], X[:,1])
plt.axis('equal') # para tener misma escala en eje x y y 
plt.show
```

#Clculo de covarianza




```python
X_mean = np.mean(X, axis= 0)
X_mean 
```

```python
cov_matrix = (X - X_mean).T.dot((X - X_mean))/(X.shape[0]-1) # .shape para ver la dimension 
```

```python
np.cov(X.T)
```

```python
eig_vals, eig_vecs = np.linalg.eig(cov_matrix)
eig_vals, eig_vecs
```

 # Calculo de matriz de correlacion y valores 


```python
corr_matrix = np.corrcoef(X.T)
corr_matrix
```

```python
eig_vals_corr, eig_vecs_corr = np.linalg.eig(corr_matrix)
eig_vals_corr, eig_vecs_corr
```

```python
total_sum=np.sum(eig_vals)
val_porc_eig_val_1= eig_vals[0]/total_sum*100
val_porc_eig_val_2= eig_vals[1]/total_sum*100
val_porc_eig_val_1, val_porc_eig_val_2
```

```python
# ordenando valores y vectores propios de mayor a menor 
eig_pairs = [(eig_vals[i], eig_vecs[i]) for i in range(len(eig_vals))]
```

```python
eig_pairs.sort(reverse=True)
eig_pairs
```

```python
total_sum = np.sum(eig_vals)
porc_eig = [eig_val/total_sum * 100 for eig_val, _ in eig_pairs]
cum_porc_eig = np.cumsum(porc_eig)
porc_eig, cum_porc_eig
```

```python
# plots 
x = ['Val prop 1', 'Val prop 2']
plt.bar(x, porc_eig)
plt.plot(x, cum_porc_eig, '.-', c="orange", linewidth=5, markersize=20)
# la recta amarilla es un valor de tolerancia que indica que tanto estas dispuesto a perder info 
```
