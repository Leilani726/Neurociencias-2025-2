```python
X = df.iloc[:, :4].values
X_std = (X - np.mean(X, axis=0)) / np.std(X, axis=0)
```

```python
# Crear un DataFrame con los datos normalizados
df_normalized = pd.DataFrame(X_std, columns=df.columns[:4])

# Calcular la matriz de covarianza
cov_matrix = df_normalized.cov()
```

```python
# Calcular los valores y vectores propios de la matriz de covarianza
eig_vals, eig_vecs = np.linalg.eig(cov_matrix)

# Calcular el porcentaje de varianza explicada por cada componente principal
explained_variance = eig_vals / np.sum(eig_vals)
```

```python
# Verificar si el 85% de la varianza est explicada
cumulative_variance = np.cumsum(explained_variance)
var_85 = cumulative_variance >= 0.85
```

```python
# Mostrar los resultados
print("Datos normalizados:")
print(df_normalized.head())
print("\nMatriz de covarianza:")
print(cov_matrix)
print("\nPorcentaje de varianza explicada por cada componente principal:")
for i, variance in enumerate(explained_variance):
    print(f"Componente principal {i+1}: {variance*100:.2f}%")
print("\nEl 85% de la varianza est explicada?:", var_85)

# Graficar los datos normalizados
plt.scatter(X_std[:, 0], X_std[:, 1])
plt.xlabel(df.columns[0])
plt.ylabel(df.columns[1])
plt.title('Datos Normalizados')
plt.show()
```
