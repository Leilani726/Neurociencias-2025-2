```python
#%% md
# Tarea Melannie Martnez
#%%
import os
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

# Cargar el dataset
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['target'] = iris.target

X, Y = df.iloc[:, :-1], df.iloc[:, -1]

# rbol con profundidad mxima de 3
clf = DecisionTreeClassifier(criterion='entropy', max_depth=3, random_state=99)
clf.fit(X, Y)

# Visualizar el rbol
plt.figure(figsize=(12,8))
plot_tree(clf, filled=True, feature_names=df.columns[:-1], class_names=iris.target_names)
plt.title("rbol de Decisin (max_depth=3)")
plt.show()

# Prediccin y matriz de confusin
observado = clf.predict(X)
cm = confusion_matrix(Y, observado)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=iris.target_names)
disp.plot(cmap=plt.cm.Blues)
plt.title("Matriz de Confusin")
plt.show()
```
