```python
import os
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.model_selection import KFold
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
```

```python
path = r"C:\Users\holai\OneDrive\Escritorio\Semestre 6\Modelos Computacionales II (F)\Neurociencias-2025-2\S03_datasets\iris\iris.csv"
df = pd.read_csv(path)
```

```python
df.value_counts()
```

```python
df.describe(include = "all") # análisis descriptivo de los datos
```

```python
X, y = df.iloc[:, :-1], df.iloc[:, -1]
```

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree # usamos el classifier si queremos clasificar los tipos de flores / el regressor para una regresión
```

```python
clf = DecisionTreeClassifier(criterion = "entropy", random_state = 99, max_depth=3)
clf.fit(X, y)
```

```python
plt.figure(figsize = (12, 8))
plot_tree(clf, filled=True, feature_names = df.columns[:-1], class_names = df["Species"].unique())
plt.show()
```

```python
y_pred = clf.predict(X)
```

```python
cm = confusion_matrix(y, y_pred, labels=clf.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=clf.classes_)
disp.plot(cmap="Blues")
plt.title("Matriz de Confusión (entrenamiento completo)")
plt.show()
```

```python
clf1 = DecisionTreeClassifier(criterion = "entropy", random_state = 99)
clf1.fit(X, y)
```

```python
plt.figure(figsize = (12, 8))
plot_tree(clf1, filled=True, feature_names = df.columns[:-1], class_names = df["Species"].unique())
plt.show()
```

```python
y1_pred = clf1.predict(X)
```

```python
cm1 = confusion_matrix(y, y1_pred, labels=clf1.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix=cm1, display_labels=clf1.classes_)
disp.plot(cmap="Blues")
plt.title("Matriz de Confusión (Árbol sin podar)")
plt.show()
```
