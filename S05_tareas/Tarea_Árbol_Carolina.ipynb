```python
# TAREA 29 de abril 2025
# rboles de clasificacin 
import os
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

#rbol podado
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['target'] = iris.target

X, Y = df.iloc[:, :-1], df.iloc[:, -1]

clf = DecisionTreeClassifier(criterion='entropy', random_state=99, max_depth=3) # max_depth pone el nivel mximo al que crecer el rbol
clf.fit(X, Y)

plt.figure(figsize=(12,8))
plot_tree(clf, filled=True, feature_names=df.columns[:-1], class_names=iris.target_names)
plt.show()

observado = clf.predict(X)
print(pd.crosstab(Y, observado))
```

```python
# MATRIZ DE CONFUSIN rbol podado
matriz = confusion_matrix(Y, observado)
disp = ConfusionMatrixDisplay(confusion_matrix=matriz, display_labels=iris.target_names)

disp.plot(cmap='Blues')
plt.title("Matriz de Confusin del rbol Podado")
plt.show()
```

```python
#rbol sin podar
klf = DecisionTreeClassifier(criterion='entropy', random_state=99, max_depth=None)
klf.fit(X, Y)

plt.figure(figsize=(12,8))
plot_tree(klf, filled=True, feature_names=df.columns[:-1], class_names=iris.target_names)
plt.show()

obsrvd = klf.predict(X)
print(pd.crosstab(Y, obsrvd))
```

```python
# MATRIZ DE CONFUSIN rbol sin podar
matrizq = confusion_matrix(Y, obsrvd)
dispq = ConfusionMatrixDisplay(confusion_matrix=matrizq, display_labels=iris.target_names)

dispq.plot(cmap='Blues')
plt.title("Matriz de Confusin del rbol Original (sin podar)")
plt.show()
```
